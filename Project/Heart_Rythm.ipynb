```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mitbih_train = pd.read_csv("mitbih_train.csv")
mitbih_train.columns = list(range(len(mitbih_train.columns)))
mitbih_train.head(5)
```

```python
mitbih_test = pd.read_csv("mitbih_test.csv")
mitbih_test.columns = list(range(len(mitbih_test.columns)))
mitbih_test.head(5)
```

```python
# 5 classes
np.unique(mitbih_train.iloc[:,-1])
```

```python
labels = ["Normal-0","S-1", "V-2", "F-3", "Q-4"]
color = ["red", "green", "black", "blue", "orange"]
fig, axs = plt.subplots(5,1, sharex = True, figsize = (15,12))
for i, j in enumerate(axs):
    j.plot((mitbih_train[mitbih_train.iloc[:,-1] == i].iloc[0])[:-1], label = labels[i], color = color[i])
    j.legend(fontsize = 13)
fig.text(0.5, 0.04, 'Time (ms)', va='center', ha='center', fontsize=15)
fig.text(0.04, 0.5, 'Normalized value', va='center', ha='center', rotation='vertical', fontsize=15)
plt.show()
```
